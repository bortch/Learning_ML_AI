# Classification des Iris

Le dataset est fourni dans la librairie Scikit Learn.

[iris Dataset](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html#sklearn.datasets.load_iris)

````python
from sklearn.datasets import load_iris

iris = load_iris()

print("Catégorie:\n{}".format(iris.target_names))
````

````python
X = iris.data
y = iris.target
print("X: {}\ny: {}".format(X.shape, y.shape))
````

````python
import pandas as pd
import numpy as np

# On veut afficher l'ensemble des valeurs du dataset
pd.set_option('display.max_rows',None)

# on concatène les features et la target 
Xy = np.c_[X,y] # c_ : concatène selon un axe

Xy
````

````python
# Création du dataframe
df = pd.DataFrame(Xy, columns=iris.feature_names+["target"])

df
````

````python
# transformer les valeurs des targets en donnée interpretable
# on les transforme en catégorie

df['target_cat']=df.target.astype("category")
print("Types des features:\n{}".format(df.dtypes))

df['target_cat']=df.target_cat.cat.rename_categories(iris.target_names)

df
````

````python
def species(fit_model, sepal_length, sepal_width, petal_length, petal_width):
    #format input as vector of feature's value
    x = np.array([sepal_length, sepal_width, petal_length, petal_width]).reshape(1,4)
    prob = fit_model.predict_proba(x)
    result = np.c_[iris.target_names,prob[0]]
    print("It's a {}!".format(iris.target_names[fit_model.predict(x)][0]))
    print("estimation:\n{}".format(result))
    
````

Pour éviter d'écrire une boucle `for`, on peut utiliser la fonction `validation_curve` qui teste toutes les valeurs pour un hyper-paramètre donné

C'est une sorte de `GridSearchCV` mais pour l'évaluation d'un unique paramètre.

#### Grid Search
La validation_curve ne permet seulement que de règler un hyper-paramètre à la fois. 

`GridSearchCV` permet de règler plusieurs hyper-paramètres simultanément pour améliorer la performance du modèle.
La méthode évalue toutes les combinaisons de plusieurs hyper-paramètres et donne la combinaison qui débouche sur la meilleure performance.

```python
# Signature
grid = GridSearchCV(model, param_grid, cv)
```

`GridSearchCV` construit une grille de modèles avec toutes les combinaisons d'hyper-paramètres présents dans `param_grid`

````python
from sklearn.model_selection import GridSearchCV
````

````python
param_grid ={'n_neighbors': np.arange(1,20),
             'metric' : ['euclidean', 'manhattan', 'minkowski']}

grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv = 5)

grid.fit(X_train, y_train)
````

#### Courbe d'apprentissage

> Est-ce que mon modèle pourrait-être plus performant si on lui donnait plus de données? 

Au moyens des courbes d'apprentissage (learning curve) où on voit la performance d'un modèle en fonction des quantités de données qui lui sont fournies.

Celle-ci est utile car à un certain moment, l'augmentation de la performance du modèle dû à l'augmentation du nombre de données fournies finit par atteint un plafond. 

On sait alors qu'il convient d'optimiser le modèle par d'autres approches et on peut arrêter la récolte des données qui est parfois coûteuse.


````python
from sklearn.model_selection import learning_curve

N, train_score_learning, val_score_learning = learning_curve(
    knn_fit, X_train, y_train, 
    train_sizes = np.linspace(0.2, 1, 5), cv = 5)

print(N)
````

Ces différentes valeurs sont les pourcentages de lots d'entrainement. 

```python
np.linspace(0.2, 1, 5)
```
```raw
>> array([0.2, 0.4, 0.6, 0.8, 1. ])
```
Dans l'exemple ci-dessus, pour 100 données, on a 20, 40, 60, 80, 100 % des données qui vont être utilisées chacun à leur tour pour l'entrainement

````python
plt.plot(N, val_score_learning.mean(axis = 1), label = 'validation')
plt.plot(N, train_score_learning.mean(axis = 1), label = 'train')

plt.ylabel('score')
plt.xlabel('train_sizes')
plt.legend()
````
