# Classification des Iris

Le dataset est fourni dans la librairie Scikit Learn.

[iris Dataset](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html#sklearn.datasets.load_iris)

```python
from sklearn.datasets import load_iris

iris = load_iris()

print("Catégorie:\n{}".format(iris.target_names))
```

```python
X = iris.data
y = iris.target
print("X: {}\ny: {}".format(X.shape, y.shape))
```

```python
import pandas as pd
import numpy as np

# On veut afficher l'ensemble des valeurs du dataset
pd.set_option('display.max_rows',None)

# on concatène les features et la target 
Xy = np.c_[X,y] # c_ : concatène selon un axe

Xy
```

```python
# Création du dataframe
df = pd.DataFrame(Xy, columns=iris.feature_names+["target"])

df
```

```python
# transformer les valeurs des targets en donnée interpretable
# on les transforme en catégorie

df['target_cat']=df.target.astype("category")
print("Types des features:\n{}".format(df.dtypes))

df['target_cat']=df.target_cat.cat.rename_categories(iris.target_names)

df
```

```python
import seaborn as sbn

sbn.pairplot(df.drop(['target'],axis=1),hue="target_cat", corner=True)

# sur la diagonale, on voit la fréquence en y 
# et la valeur de la feature en x

# il y a une corrélation positive entre petal width et petal length
```

```python
df.describe()
```

# Train & Test Sets

```python
# récupération des valeurs entières de chaque catégorie
df["target"]=df.target_cat.cat.codes
```

```python
from sklearn.model_selection import train_test_split as tts

X = df.drop(['target','target_cat'], axis=1)
y = df.target

X_train, X_test, y_train, y_test = tts(X,y, test_size = 0.25, random_state=0)
# split 25% de test et 75% de train

print("Train set: {}\nTest set: {}\n".format(X_train.shape, X_test.shape))
```

# Machine Learning Models

## 1) Linear Model (OLS)

```python
from sklearn.linear_model import LinearRegression

# on utilise les paramètres par défaut
linear_model = LinearRegression()
# on entraine le model
lmf = linear_model.fit(X_train, y_train)
lmf.predict(X_test)
```

```python
lmf.score(X_test,y_test) 
# coefficient de détermination qui nous dit dans 
# quelle mesure notre droite
```

```python
lmf.coef_
# 4 dimensions car 4 features
# le signe des coefficients montre une diminution ou augmentation
# pour chaque feature pour lors du passage d'une catégorie (classe) à une autre
# (classe: setosa | versicolor | verginica)
# l'interpretation des coefficients dans le cas d'une classification n'est pas
# vraiment possible.
```

## 2) Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(max_depth=4, random_state=0)

dtc_fit=dtc.fit(X_train,y_train)
```

```python
dtc_fit
```

```python
# visualisation de l'arbre de décision

from sklearn import tree
from datetime import datetime
from matplotlib.pyplot import figure
%matplotlib inline
figure(figsize=(10, 8), dpi=80)
now = datetime.utcnow()
tree.plot_tree(dtc_fit,feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.savefig("dtc_"+str(dtc_fit.get_depth())+'_'+now.strftime("%Y%d%m_%H_%M_%S")+".pdf")
plt.show
```

### Score & Accuracy

```python
y_pred = dtc.predict(X_test)
```

```python
dtc.score(X_test,y_test) 
```

## Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(y_test, y_pred)

conf_matrix = pd.DataFrame(confusion_matrix(y_test, y_pred), columns = iris.target_names, index = iris.target_names)
conf_matrix
```

```python
# Beautifull Matrix confusion using Heatmap

sbn.heatmap(conf_matrix, annot=True, cbar=False, cmap='Blues_r')
```

**Attention**

- `index` = *valeurs réelles*
- `columns` = *predictions*

Dans l'exemple, on prédit comme une fleur comme étant de la classe `virginica` alors qu'il s'agit d'une `versicolor`

```python
# Exemple érronément catégorisé
def getFailPredictionInstance(y_test, y_pred):
    df['fail']=(y_test==y_pred)
    print(df[df['fail']==False])
    
getFailPredictionInstance(y_test, y_pred)
```

```python
def species(fit_model, sepal_length, sepal_width, petal_length, petal_width):
    #format input as vector of feature's value
    x = np.array([sepal_length, sepal_width, petal_length, petal_width]).reshape(1,4)
    prob = fit_model.predict_proba(x)
    result = np.c_[iris.target_names,prob[0]]
    print("It's a {}!".format(iris.target_names[fit_model.predict(x)][0]))
    print("estimation:\n{}".format(result))
    
```

```python
species(dtc_fit, 6,2.7,4.5,1.6)
```
